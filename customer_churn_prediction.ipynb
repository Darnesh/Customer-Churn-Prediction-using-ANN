```python
data.select_dtypes(include=['object'])
```

```python
def uniq(data):
    return data.unique()

data.select_dtypes(include=['object']).apply(uniq, axis=0)
```

```python
for column in data.columns:
    if data[column].dtypes==object:
        print(f"{column} : {data[column].unique()}")
```

```python
data.replace(['No phone service','No internet service'],'No',inplace=True)
```

```python
data.select_dtypes(include=['object']).apply(uniq, axis=0)
```

```python
data.select_dtypes(include=['object']).columns
```

```python
data = pd.get_dummies(data, columns=data.select_dtypes(include=['object']).columns,drop_first=True)
```

```python
data.head() 
```

```python
scale_cols = ['tenure','MonthlyCharges','TotalCharges']
from  sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[scale_cols] = scaler.fit_transform(data[scale_cols])
```

```python
data.sample(5)
```

```python
data.rename(columns={'Churn_Yes':'churn'})
```

```python
x = data.iloc[::,0:-1]
y = data.iloc[::,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train.columns)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(input_dim=23, units=15, activation='relu'))
model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_history = model.fit(x_train, y_train, validation_split=0.40, batch_size=100, epochs=100)
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```

```python
test_loss , test_accuracy
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_pred[y_pred>0.5] = 1
y_pred[y_pred==0.5] = 1
y_pred[y_pred<0.5] = 0
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
```
